```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('12cdenhanced.csv', low_memory=False)
```

```python
df.columns.values
```

```python
df['reg_datetime'] = df['Voters_OfficialRegDate'].apply(lambda x: datetime.datetime.strptime(x, '%m/%d/%Y'))
```

```python
def label_newVoter1(row):
    # checks both 2016 and 2010 primaries and registered before 2010 primary
    if row['Primary_2018-06-26'] == 'Y' and row['Primary_2016-06-28']!='Y' and row['Primary_2010-09-14']!='Y' and \
    row['reg_datetime'] < datetime.datetime.strptime('9/14/2010', '%m/%d/%Y'):
        return 1
    else:
        return 0

def label_newVoter2(row):
    # checks only 2016 and 2010 primaries and registered before 2016 primary
    if row['Primary_2018-06-26'] == 'Y' and row['Primary_2016-06-28']!='Y' and row['Primary_2010-09-14']!='Y' and \
    row['reg_datetime'] < datetime.datetime.strptime('6/28/2016', '%m/%d/%Y'):
        return 1
    else:
        return 0
    
df['NewToLastPrimary'] = df.apply(lambda x: label_newVoter1(x), axis=1)
print('1\n', df['NewToLastPrimary'].value_counts())
df['NewToLastPrimary'] = df.apply(lambda x: label_newVoter2(x), axis=1)
print('2\n', df['NewToLastPrimary'].value_counts())
```
