```python
import pandas as pd
```

```python
df = pd.read_csv('12cdenhanced.csv', low_memory=False)
```

```python
df.head()
```

```python
LatestPrimary = 'Primary_2018-06-26'
PastPrimaries = ['Primary_2016-06-28', 'Primary_2010-09-14']
```

```python
def label_newVoter(row):
    if row[LatestPrimary] == 'Y' and row[PastPrimaries[0]]!='Y' and row[PastPrimaries[1]]!='Y':
        return 'NewVoter'
    else:
        return 'NotNewVoter'
```

```python
df['NewToLastPrimary'] = df.apply(lambda row: label_newVoter(row), axis=1)
```

```python
df[['Primary_2018-06-26', 'Primary_2016-06-28', 'Primary_2010-09-14', 'NewToLastPrimary']]
```

```python
new_voter_df = df[df['NewToLastPrimary']=='NewVoter']
not_new_voter_df = df[df['NewToLastPrimary']=='NotNewVoter']
```

```python
def key_getter(item):
    return item[-2:]+item[-6:-3]

df['key'] = df['Precinct'].apply(lambda x: key_getter(x))
```

```python
df['key']
```

```python
df.groupby(by = 'key').Voters_Age.median()
```

```python
df.groupby(by = 'key', as_index = False).count()[['key', 'NewToLastPrimary']].rename(columns = {'key':'KEY'})
```
